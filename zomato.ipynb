```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Zomato_data.csv")
df.head()
```

```python
def handleRate(value):
  value = str(value).split('/')
  value = value[0];
  return float(value)
```

```python
df['rate'] = df['rate'].apply(handleRate)
df
```

```python
df.info()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel('type of restaurant')
```

```python
gd = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':gd})
plt.plot(result,c="green",marker='o')
plt.xlabel('type of restaurant',c="blue")
plt.ylabel('votes',c="blue")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(df['rate'], bins=5, edgecolor='black')
plt.title('Distribution of Restaurant Ratings')
plt.show()
```

```python
couple_data = df['approx_cost(for two people)']
sns.countplot(x=couple_data)
plt.xlabel('approx spend by couples')
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=df)
```

```python
pt = df.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pt,annot=True,cmap='YlGnBu',fmt='d')
plt.title("heatmap")
plt.xlabel("online order")
plt.ylabel("type of restaurant")
plt.show()
```
